### Get data from Billboard.com

```python
# store the url
url = "https://www.billboard.com/charts/hot-100"
```

```python
#check status code
response = requests.get(url)
response.status_code # 200 status code means OK!
```

```python
#parsing the html code
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
soup.select('body')
```

```python
soup.find_all("span")
```

```python
soup.find_all("span",class_="chart-element__information__song text--truncate color--primary")
```

```python
#Create columns list
```

```python
song_title = []
artist = []
rank_change= []
rank_last_week = []
peak_rank = []
weeks_on_rank = []
```

```python
num_iter = len(soup.find_all("span",class_="chart-element__information__song text--truncate color--primary"))

# iterate through the result set and retrive all the data")[i].get_text())
for i in range(num_iter):
    song_title.append(soup.find_all("span",class_="chart-element__information__song text--truncate color--primary")[i].get_text())
    artist.append(soup.find_all("span",class_="chart-element__information__artist text--truncate color--secondary")[i].get_text())
    rank_change.append(soup.find_all("span", class_="chart-element__information__delta__text text--default")[i].get_text())
    rank_last_week.append(soup.find_all("span",class_="chart-element__information__delta__text text--last")[i].get_text())
    peak_rank.append(soup.find_all("span", class_="chart-element__information__delta__text text--peak")[i].get_text())
    weeks_on_rank.append(soup.find_all("span", class_ = "chart-element__information__delta__text text--week")[i].get_text())
# iterate through the result set and retrive all the data")[i].get_text())
```

```python
# construct the dataframe
```

```python
billboard_df= pd.DataFrame({"title":song_title,
                       "artist":artist,
                       "rank_change": rank_change,
                       "rank_last_week": rank_last_week,
                       "peak_rank":peak_rank,
                     "weeks_on_rank":weeks_on_rank
                      })

billboard_df.head()
```
